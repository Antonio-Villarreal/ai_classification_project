```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

### Load Dataset

```python
realvsfake = os.path.join('dataset', 'real_vs_fake', 'real-vs-fake')
```

```python
subdirectories = ['real', 'fake']
test = os.path.join(realvsfake, 'test')
train = os.path.join(realvsfake, 'train')
valid = os.path.join(realvsfake, 'valid')
```

```python
def create_subdirs(directory):
    dict = {}
    for subdir in subdirectories:
        dict[subdir] = os.path.join(directory, subdir)
    return dict
```

```python
testdirs = create_subdirs(test)
traindirs = create_subdirs(train)
validdirs = create_subdirs(valid)
```

## Real Faces

```python
def display_batch(directory, start):
    fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(15, 5))
    for i in range(4):
        image_path = os.path.join(directory, f"000{i+start}.jpg")
        image = Image.open(image_path)
        img = image.resize((256, 256))
        
        axes[i].imshow(img)
        axes[i].axis('off')
        
    plt.tight_layout()
    plt.show()
```

```python
directory = traindirs['real']
display_batch(directory, 48)
```

```python
display_batch(directory, 81)
```

## Fake Faces

```python
def display_img(directory, img_name):
    img_path = os.path.join(directory, img_name)
    img = plt.imread(img_path)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```
